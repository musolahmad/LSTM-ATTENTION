```python
import pandas as pd
# Use pd.read_json to read data from the JSON file
df_a = pd.read_json("tinggi_air.json")
df_filter = df_a[['ID_PINTU_AIR','NAMA_PINTU_AIR','LATITUDE','LONGITUDE','SIAGA1','SIAGA2','SIAGA3']]
df_filter
```

```python
import requests
from datetime import datetime
import pytz

# Fungsi untuk mendapatkan ketinggian menggunakan Open Elevation API
def get_elevation(latitude, longitude):
    url = f"https://api.open-elevation.com/api/v1/lookup?locations={latitude},{longitude}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        elevation = data['results'][0]['elevation']
        return elevation
    else:
        return None

df_filter['elevation'] = df_filter.apply(lambda row: get_elevation(row['latitude'], row['longitude']), axis=1)
df_filter
```

```python
from geopy.distance import geodesic

# Create a list to store the results
distance_data = []

# Iterate through each row in df_filter
for index_filter, row_filter in df_filter.iterrows():
    gate_id = row_filter['id_pintu_air']
    gate_name = row_filter['nama_pintu_air']
    gate_lat = row_filter['latitude']
    gate_lon = row_filter['longitude']
    gate_elv = row_filter['elevation']
    gate_b = row_filter['bahaya']
    gate_s = row_filter['siaga']
    gate_w = row_filter['waspada']
    gate_coords = (gate_lat, gate_lon)

    # Iterate through each row in df
    for index_df, row_df in df.iterrows():
        location_id = row_df['id']
        location_name = row_df['lokasi']
        location_lat = row_df['latitude']
        location_lon = row_df['longitude']
        location_elv = row_df['elevation']
        location_coords = (location_lat, location_lon)

        # Calculate the geodesic distance (Haversine) in kilometers
        distance_km = geodesic(gate_coords, location_coords).km

        # Append the results to the list
        distance_data.append({
            'id_pintu_air': gate_id,
            'nama_pintu_air': gate_name,
            'tinggi_lokasi_pintu': gate_elv,
            'bahaya': gate_b,
            'waspada': gate_w,
            'siaga': gate_s,
            'id_lokasi': location_id,
            'nama_lokasi': location_name,
            'tinggi_lokasi_evakuasi': location_elv,
            'distance_km': distance_km
        })

# Create a new DataFrame from the distance data
df_distances_geodesic = pd.DataFrame(distance_data)

# Display the new DataFrame
display(df_distances_geodesic.head())
```

```python
import pandas as pd
import osmnx as ox
import geopandas as gpd
from shapely import wkt

# Load node data from CSV
nodes_loaded = pd.read_csv('jakarta_nodes_with_elevation_final.csv', index_col=0)

# Load edge data from CSV
edges_loaded = pd.read_csv('jakarta_edges.csv', index_col=[0, 1, 2]) # Index is u, v, key

# Convert 'geometry' column from string to shapely.geometry objects if it exists
if 'geometry' in nodes_loaded.columns and nodes_loaded['geometry'].notna().any():
    nodes_loaded['geometry'] = nodes_loaded['geometry'].apply(wkt.loads)
else:
    # If geometry is missing or all NaN, create Point geometries from x, y
    nodes_loaded['geometry'] = gpd.points_from_xy(nodes_loaded['x'], nodes_loaded['y'])

if 'geometry' in edges_loaded.columns and edges_loaded['geometry'].notna().any():
    edges_loaded['geometry'] = edges_loaded['geometry'].apply(wkt.loads)

# Convert to GeoDataFrames with a standard geographical CRS (EPSG:4326)
gdf_nodes = gpd.GeoDataFrame(nodes_loaded, geometry='geometry', crs='EPSG:4326')
gdf_edges = gpd.GeoDataFrame(edges_loaded, geometry='geometry', crs='EPSG:4326')

# Reconstruct the graph from the GeoDataFrames
G_reconstructed = ox.graph_from_gdfs(gdf_nodes, gdf_edges)
```

```python
def calculate_fuzzy_weight(tta, d_max):
    # 1. Menentukan Skor Risiko (rs) - Tabel III.7 [cite: 601, 602]
    if d_max <= 0.05: rs = 0
    elif d_max <= 0.10: rs = 1
    elif d_max <= 0.20: rs = 2
    elif d_max <= 0.30: rs = 3
    elif d_max <= 0.50: rs = 4
    elif d_max <= 1.00: rs = 5
    else: rs = 999999 
    
    # 2. Risk Penalty (rp) = rs * 5 menit [cite: 609]
    rp = rs * 5
    
    # 3. Parameter TFN (a1, a2, a3, a4) - Tabel III.8 [cite: 610-615]
    a1, a2, a3 = 0.8 * tta, 0.9 * tta, 1.1 * tta
    a4 = (1.5 * tta) + rp
    
    # 4. Cost Crisp (Pc) - Persamaan III.1 [cite: 618]
    pc = (a1 + 2*a2 + 2*a3 + a4) / 6
    return pc
```

```python
import networkx as nx
import osmnx as ox
import pandas as pd
from scipy.spatial import distance  # BARIS INI WAJIB ADA

def cari_rute_evakuasi_cerdas(id_pintu, prediksi_tma_cm, G):
    # 1. Ambil Metadata Pintu
    info = pintu_air_db.get(id_pintu)
    if not info: return "ID Pintu tidak ditemukan"

    # 2. Hitung Water Level Absolut [cite: 593, 598]
    # Selisih prediksi terhadap batas normal (cm ke meter)
    h_flood_m = max(0, (prediksi_tma_cm - info['waspada']) / 100)
    water_level_abs = info['elevation'] + h_flood_m
    
    print(f"Menganalisis: {info['nama_pintu_air']} | elevasi : {info['elevation']} | Status Luapan: {h_flood_m}m")

    # 3. Update Bobot Fuzzy di Setiap Edge [cite: 589, 592]
    for u, v, k, data in G.edges(data=True, keys=True):
        tta = data.get('travel_time', 1.0)
        
        # Ambil elevasi jalan (rata-rata node u dan v)
        elev_u = G.nodes[u].get('elevation', 0)
        elev_v = G.nodes[v].get('elevation', 0)
        elev_jalan = (elev_u + elev_v) / 2
        
        # D_max: Jika air lebih tinggi dari aspal, ada genangan [cite: 596, 602]
        d_max = max(0, water_level_abs - elev_jalan)
        
        # Hitung Pc (Cost Crisp)
        data['fuzzy_weight'] = calculate_fuzzy_weight(tta, d_max)

    # 4. Cari Lokasi Evakuasi Terdekat secara Spasial [cite: 584, 586]
    pintu_pos = (info['latitude'], info['longitude'])
    target_terdekat = min(evakuasi_db, key=lambda x: distance.euclidean(pintu_pos, (x['latitude'], x['longitude'])))
    
    # 5. Proyeksi ke Node Graf
    node_source = ox.nearest_nodes(G, X=info['longitude'], Y=info['latitude'])
    node_target = ox.nearest_nodes(G, X=target_terdekat['longitude'], Y=target_terdekat['latitude'])
    
    # 6. Jalankan Fuzzy Dijkstra [cite: 320, 625]
    try:
        path = nx.shortest_path(G, source=node_source, target=node_target, weight='fuzzy_weight')
        print(f"Rute Aman ke '{target_terdekat['lokasi']}' berhasil dibuat. elevasi '{target_terdekat['elevation']}' ")
        return path, target_terdekat['lokasi']
    except nx.NetworkXNoPath:
        return None, "Jalur Terputus Total!"
```

```python
def cari_rute_evakuasi_target(id_pintu, prediksi_tma_cm, G):
    # 1. Ambil Metadata Pintu
    info = pintu_air_db.get(id_pintu)
    if not info: return None, "ID Pintu tidak ditemukan", None

    # 2. Hitung Water Level Absolut (mdpl)
    h_flood_m = max(0, (prediksi_tma_cm - info['waspada']) / 100)
    water_level_abs = info['elevation'] + h_flood_m
    
    print(f"Menganalisis: {info['nama_pintu_air']} | Elevasi Pintu: {info['elevation']} | Status Luapan: {h_flood_m}m")
    print(f"Water Level Banjir: {water_level_abs:.2f} mdpl")

    # 3. Update Bobot Fuzzy di Setiap Edge (Wajib dilakukan sebelum mencari target)
    for u, v, k, data in G.edges(data=True, keys=True):
        tta = data.get('travel_time', 1.0)
        elev_u = G.nodes[u].get('elevation', 0)
        elev_v = G.nodes[v].get('elevation', 0)
        elev_jalan = (elev_u + elev_v) / 2
        
        d_max = max(0, water_level_abs - elev_jalan)
        data['fuzzy_weight'] = calculate_fuzzy_weight(tta, d_max)

    # 4. Cari Lokasi Evakuasi Berbasis Jaringan (Menangani sungai/hambatan)
    node_source = ox.nearest_nodes(G, X=info['longitude'], Y=info['latitude'])
    
    # Hitung semua biaya perjalanan (fuzzy) dari titik bencana ke seluruh graf
    weight_type = 'fuzzy_weight'
    all_path_costs = nx.single_source_dijkstra_path_length(G, node_source, cutoff=5000, weight=weight_type)

    kandidat_aman = []
    for loc in evakuasi_db:
        # Syarat 1: Lokasi harus lebih tinggi dari air (Aman)
        if loc['elevation'] > water_level_abs:
            node_loc = ox.nearest_nodes(G, X=loc['longitude'], Y=loc['latitude'])
            
            # Syarat 2: Lokasi harus terjangkau oleh jaringan jalan
            if node_loc in all_path_costs:
                kandidat_aman.append({
                    'node_id': node_loc,
                    'cost': all_path_costs[node_loc], # Jarak tempuh nyata lewat jalan
                    'data': loc
                })
    
    if not kandidat_aman:
        return None, "Gagal: Tidak ada lokasi evakuasi aman yang terjangkau!", None

    # Pilih target berdasarkan COST JARINGAN terkecil (Bukan Jarak Burung)
    target_pilihan = min(kandidat_aman, key=lambda x: x['cost'])
    target_terdekat = target_pilihan['data']
    node_target = target_pilihan['node_id']
    
    # 5. Jalankan Fuzzy Dijkstra ke Target Terpilih
    try:
        path = nx.shortest_path(G, source=node_source, target=node_target, weight=weight_type)
        total_cost = target_pilihan['cost']
        
        print(f"Rute Aman ke '{target_terdekat['lokasi']}' (Elevasi: {target_terdekat['elevation']})")
        print(f"Total Fuzzy Cost (Waktu + Risiko): {total_cost:.2f}")

        # 6. Ekstraksi Detail Koordinat untuk Tabel
        route_details = []
        for i, node_id in enumerate(path):
            node_data = G.nodes[node_id]
            label = f"AWAL: {info['nama_pintu_air']}" if i == 0 else \
                    f"TUJUAN: {target_terdekat['lokasi']}" if i == len(path)-1 else f"Titik {i}"
            
            route_details.append({
                'label': label,
                'lat': node_data['y'],
                'lon': node_data['x'],
                'elevation': node_data.get('elevation', 0)
            })
            
        return path, route_details, target_terdekat['lokasi']
        
    except nx.NetworkXNoPath:
        return None, "Jalur Terputus Total!", None
```

```python
def cari_rute_standar(id_pintu, G):
    # 1. Ambil Metadata Pintu Air
    info = pintu_air_db.get(id_pintu)
    if not info: 
        return None, "ID Pintu tidak ditemukan"

    print(f"Mencari rute standar dari: {info['nama_pintu_air']}")

    # 2. Cari Lokasi Evakuasi Terdekat Secara Spasial [cite: 584, 586]
    pintu_pos = (info['latitude'], info['longitude'])
     # Menentukan bobot yang digunakan (misal: waktu tempuh)
    weight_type = 'travel_time' 
    # 3. Proyeksi Koordinat ke Node ID di Graf [cite: 584, 586]
    node_source = ox.nearest_nodes(G, X=info['longitude'], Y=info['latitude'])

    all_path_costs = nx.single_source_dijkstra_path_length(G, node_source, cutoff=5000, weight=weight_type)

    kandidat_aman = []
    for loc in evakuasi_db:
        node_loc = ox.nearest_nodes(G, X=loc['longitude'], Y=loc['latitude'])
        # Syarat 2: Lokasi harus terjangkau oleh jaringan jalan
        if node_loc in all_path_costs:
           kandidat_aman.append({
              'node_id': node_loc,
              'cost': all_path_costs[node_loc], # Jarak tempuh nyata lewat jalan
              'data': loc
           })
    if not kandidat_aman:
        return None, "Gagal: Tidak ada lokasi evakuasi aman yang terjangkau!", None

    # Pilih target berdasarkan COST JARINGAN terkecil (Bukan Jarak Burung)
    target_pilihan = min(kandidat_aman, key=lambda x: x['cost'])
    target_terdekat = target_pilihan['data']
    node_target = target_pilihan['node_id']
    
    # 4. Eksekusi Dijkstra Standar
    # Menggunakan 'travel_time' atau 'length' sebagai bobot statis [cite: 590, 591]
    try:        
        path = nx.shortest_path(G, 
                                source=node_source, 
                                target=node_target, 
                                weight=weight_type)
        
        # Menghitung total biaya (waktu/jarak) rute tersebut [cite: 313]
        total_cost = nx.shortest_path_length(G, source=node_source, target=node_target, weight=weight_type)
        
        print(f"Rute Standar ke '{target_terdekat['lokasi']}' ditemukan.")
        print(f"Estimasi {weight_type}: {total_cost:.2f}")
        
        # 6. Ekstraksi Detail Koordinat, Elevasi, dan Nama 
        route_details = []
        for i, node_id in enumerate(path):
            node_data = G.nodes[node_id]
            label = f"AWAL: {info['nama_pintu_air']}" if i == 0 else \
                    f"TUJUAN: {target_terdekat['lokasi']}" if i == len(path)-1 else f"Titik {i}"
            
            route_details.append({
                'lat': node_data['y'],
                'lon': node_data['x'],
                'elevation': node_data.get('elevation', 0),
                'label': label
            })
            
        return path, route_details, target_terdekat['lokasi']
    
    except nx.NetworkXNoPath:
        return None, "Jalur tidak terhubung"
```
